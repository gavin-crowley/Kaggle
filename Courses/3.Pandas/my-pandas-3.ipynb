```python
# remean the price to 0
```

```python
melb_price_mean = melb.Price.mean()
melb.Price.map(lambda p: p - melb_price_mean)
```

```python
def remean_price(row):
    row.Price = row.Price - melb_price_mean
    return row

melb.apply(remean_price, axis='columns')
```

```python
# Note that map() and apply() return new, transformed Series and DataFrames, respectively. They don't modify the original data they're called on. If we look at the first row of reviews, we can see that it still has its original points value.
```

```python
# faster way
melb_price_mean = melb.Price.mean()
melb.Price - melb_price_mean
```

```python
#Another q
n_arm = melb.Suburb.map(lambda desc: "Armadale" in desc).sum()
n_haw = melb.Suburb.map(lambda desc: "Hawthorn" in desc).sum()
descriptor_counts = pd.Series([n_arm, n_haw], index=['Armadale', 'Hawthorn'])
descriptor_counts
```

```python
#Another q
def stars(row):
    if row.Distance == 'Hawthorn':
        return 3
    elif row.Distance >= 40:
        return 4
    elif row.Distance >= 45:
        return 5
    else:
        return 1

star_ratings = melb.apply(stars, axis='columns')
star_ratings
```

```python
#---------------------
#Groupwise analysis
```

```python
import pandas as pd
melb = pd.read_csv('../../../Datasets/kaggle/melb_data.csv')
melb.head()
```

```python
melb.groupby('Postcode').Postcode.count()
#also, melb.groupby('Postcode').size()
```

```python
melb.groupby('Price').Price.min()
```

```python
melb.groupby('Suburb').apply(lambda df: df.Regionname.iloc[0])
```

```python
melb.groupby(['Suburb', 'Regionname']).apply(lambda df: df.loc[df.Price.idxmin()])
```

```python
#agg(), which lets you run a bunch of different functions on your DataFrame simultaneously
```

```python
melb.groupby(['Suburb']).Price.agg([len, min, max])
```

```python
#-----------------
#Multi-indexes

#Multi-indices have several methods for dealing with their tiered structure which are absent for single-level indices. They also require two levels of labels to retrieve a value.
#However, in general the multi-index method you will use most often is the one for converting back to a regular index, the reset_index() method

address_check = melb.groupby(['Suburb', 'Regionname']).Address.agg([len])
address_check
```

```python
mi = address_check.index
type(mi)
```

```python
address_check.reset_index()
```

```python
#Sorting
```

```python
address_check = address_check.reset_index()
address_check.sort_values(by='len', ascending=True)
```

```python
address_check.sort_index()
```

```python
address_check.sort_values(by=['Suburb', 'len'])
```
